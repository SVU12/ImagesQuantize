```python

def draw_rectangle(rgb_color):
    color = f'rgb({rgb_color[0]}, {rgb_color[1]}, {rgb_color[2]})'
    html = f'<svg width="100" height="100"><rect width="100" height="100" fill="{color}" /></svg>'

    display(HTML(html))

def distanceTuple(c1,c2) -> float:
    """
    Renvoie la distance euclidienne entre deux vecteurs/tuples de même dimension
    """
    return sqrt(sum([(c1[i]-c2[i])**2 for i in range(len(c1))]))

def distanceBarycentre(c1,c2) -> float:
    """
    Renvoie la distance euclidienne entre deux vecteurs de composantes rgb de même dimension
    """
    s=0
    for i in range(len(c1)):
        s+=(c1[i][0]-c2[i][0])**2+(c1[i][1]-c2[i][1])**2+(c1[i][2]-c2[i][2])**2
    return sqrt(s)

def getPixel(x : int,y : int,px):
    """
    Renvoie le pixel de coordonnées (x,y) de l'image px
    """
    return px[x,y]


```

**Question 2)** On implémente maintenant l'ago k-means, qui va nous permettre de déterminer les k couleurs les plus représentatives de l'image. On affiche ensuite la palette de couleurs obtenue. 

```python
def naif(k_total : int, im) ->list:
    
    w,h = im.size
    px=im.load()
    coord = [getPixel(i,j,px) for i in range(w) for j in range(h)]
    assoc = {(i,j):None for i in range(w) for j in range(h)}
    #On choisit k centres réparti de manière uniforme dans l'espace des couleurs
    
    
    couleurs = [px[i,j] for i in range(w) for j in range(h)]
    step = (w*h)//k_total
    
    k_centres_old = [couleurs[i*step + step//2] for i in range(k_total)]
    #On va maintenant associer chaque pixel à un centre
    for i in range(w):
        for j in range(h):
            dist = [distanceTuple(px[i,j],k_centres_old[k]) for k in range(k_total)]
            assoc[(i,j)] = dist.index(min(dist))
    
    return k_centres_old,assoc
```

```python
def k_means(k_total: int, epsilon: float, im) -> list:
    w,h = im.size
    px=im.load()
    coord = [getPixel(i,j,px) for i in range(w) for j in range(h)]
    assoc = {(i,j):None for i in range(w) for j in range(h)}
    #On choisit k centres réparti de manière uniforme dans l'espace des couleurs
    
    
    couleurs = [px[i,j] for i in range(w) for j in range(h)]
    step = (w*h)//k_total
    
    k_centres_old = [couleurs[i*step + step//2] for i in range(k_total)]
    nbTours = 0
    while True:
        
        nbTours+=1
        k_centres = [[0,0,0] for i in range(k_total)]
        #On associe chaque pixel à un centre
        for x in range(w):
            for y in range(h):
                assoc[(x,y)]=k_centres_old.index(min(k_centres_old, key=lambda c: distanceTuple(getPixel(x,y,px),c)))
        #On calcule les nouveaux centres
        nb=[0 for i in range(k_total)]
        for i in range(w):
            for j in range(h):
                nb[assoc[(i,j)]]+=1
                for k in range(3):
                    k_centres[assoc[(i,j)]][k]+=px[i,j][k]
        for i in range(k_total):
            if nb[i]==0:
                k_centres[i]=k_centres_old[i]
            else:
                for j in range(3):
                    k_centres[i][j]=int(k_centres[i][j]/nb[i])
                k_centres[i]=tuple(k_centres[i])
        
        # #On vérifie si on a atteint la précision souhaitée
        if distanceBarycentre(k_centres,k_centres_old) < epsilon:
            break
        else:
            k_centres_old = k_centres
            
    print("Nombre de tours : ",nbTours)
    return k_centres,assoc
```

```python

def remplace(im,assoc : dict,bary : list) -> Image: #Cree une nouvelle image avec les couleurs de la palette 
    """
    Renvoie une nouvelle image avec les couleurs de la palette
    """
    w,h=im.size
    im2 = im.copy()
    px2 = im2.load()
    for x in range(w):
        for y in range(h):
            px2[x,y]
            bary[assoc[(x,y)]]
            px2[x,y]=bary[assoc[(x,y)]]
    return im2
```

```python
def score(imref : Image,im : Image):
    """
    Renvoie la distance normalisée par canal sur chaque pixel entre deux images.
    """
    pxf=imref.load()
    px=im.load()
    s=0
    w,h=im.size
    for x in range(w):
        for y in range(h):
            s+=distanceTuple(pxf[x,y],px[x,y])
    return sqrt(s/(w*h*3))


#print(score(imq,im))

#On teste avec différentes valeurs de k et d'epsilon, et on affiche les résultats dans un plot matplotlib

k = [1,2,4,8,16,32]
#mem = [k_means(i,5,im) for i in k]

#imq = [remplace(im,mem[i][1],mem[i][0]) for i in range(len(k))]

#plt.figure(figsize=(10, 6))
#plt.plot(k,[score(im,imq[i]) for i in range(len(k))],  color='r', alpha=0.6, label='Score')
#Erreur de dépassement d'indice dans les associations. Mauvaise implémentation ???! étrange.
```

```python
def generateurImageAléatoire(k:int,w:int,h:int):
    """
    Renvoie une image de taille w*h avec k couleurs aléatoires
    """
    couleurs = [(random.randint(0,255),random.randint(0,255),random.randint(0,255)) for i in range(k)]
    im = Image.new("RGB",(w,h))
    px = im.load()
    for i in range(w):
        for j in range(h):
            px[i,j]=couleurs[random.randint(0,k-1)]
    return im

```

```python
import time
def diffScore(k:int,kcol:int,w:int,h:int):
    """
    Renvoie la différence de score entre l'image originale et l'image compressée
    """
    t1=time.time()
    im = generateurImageAléatoire(kcol,w,h)
    bary,assoc = k_means(k,2,im)
    imq = remplace(im,assoc,bary)
    t2=time.time()
    imqq=im.copy()
    imqq=imqq.quantize(colors=k)
    imqq=imqq.convert("RGB")
    t3=time.time()
    return [score(im,imq),score(im,imqq),t2-t1,t3-t2]
```

**Question 7)** Prétraitement. On va appliquer un filtre de flou gaussien par convolution sur l'image, pour lisser l'image et ainsi filtrer les hautes fréquences. 

```python

def generateurMasqueGaussien(taille, sigma):
    if taille % 2 == 0:
        print("Erreur : La taille du masque doit etre impaire")
        return None
    masque=np.zeros((taille, taille), dtype=float)
    somme = 0
    for i in range(taille):
        for j in range(taille):
            masque[i][j] = np.exp(-((i-(taille-1)/2)**2 + (j-(taille-1)/2)**2) / (2*sigma**2))
            somme += masque[i][j]
    for i in range(taille):
        for j in range(taille):
            masque[i][j] /= somme  # Normalisation
    return masque



def convolutionRGB(image, masque):
    """
    Renvoie une nouvelle image convoluée par le masque 
    """
    w,h = image.size
    px = image.load()
    im = Image.new("RGB",(w,h))
    px2 = im.load()
    for i in range(w):
        for j in range(h):
            s=[0,0,0]
            for k in range(len(masque)):
                for l in range(len(masque)):
                    if i+k-len(masque)//2>=0 and i+k-len(masque)//2<w and j+l-len(masque)//2>=0 and j+l-len(masque)//2<h:
                        for c in range(3):
                            s[c]+=masque[k][l]*px[i+k-len(masque)//2,j+l-len(masque)//2][c]
            for c in range(3):
                px2[i,j]=tuple([int(s[c]) for c in range(3)])
    return im


    
```

**Question 8)** On va maintenant étudier les différents profils de couleur. On va utiliser une représentation en HSL. Chromaticité, Luminance ainsi que Teinte. C'est une représentaiton cylindrique. La variance n'est pas identique sur ces 3 coordonnées. (Chromacité axe principal, Luminance module et Teinte est un angle) 
La méthode de Clustering, avec une distance usuelle ne sera absolument pas adaptée. 

```python
def distanceHSV(c1,c2) -> float:
    h1, s1, v1 = c1
    h2, s2, v2 = c2
    d = sqrt(s1**2 + s2**2 - 2 * s1 * s2 * cos((h2 - h1) * pi / 180) + (v2 - v1)**2)
    return d

import colorsys

def k_means_HSV(k_total: int, epsilon: float, im) -> list:
    w,h = im.size
    im=im.convert('HSV')
    px=im.load()
    coord = [getPixel(i,j,px) for i in range(w) for j in range(h)]
    assoc = {(i,j):None for i in range(w) for j in range(h)}
    #On choisit k centres réparti de manière uniforme dans l'espace des couleurs
    
    
    couleurs = [px[i,j] for i in range(w) for j in range(h)]
    step = (w*h)//k_total
    
    k_centres_old = [couleurs[i*step + step//2] for i in range(k_total)]
    nbTours = 0
    while True:
        
        nbTours+=1
        k_centres = [[0,0,0] for i in range(k_total)]
        #On associe chaque pixel à un centre
        for x in range(w):
            for y in range(h):
                assoc[(x,y)]=k_centres_old.index(min(k_centres_old, key=lambda c: distanceHSV(getPixel(x,y,px),c)))
        #On calcule les nouveaux centres
        nb=[0 for i in range(k_total)]
        for i in range(w):
            for j in range(h):
                nb[assoc[(i,j)]]+=1
                for k in range(3):
                    k_centres[assoc[(i,j)]][k]+=px[i,j][k]
        for i in range(k_total):
            if nb[i]==0:
                k_centres[i]=k_centres_old[i]
            else:
                for j in range(3):
                    k_centres[i][j]=int(k_centres[i][j]/nb[i])
                k_centres[i]=tuple(k_centres[i])
        
        # #On vérifie si on a atteint la précision souhaitée
        if distanceHSV(k_centres[0],k_centres_old[0]) < epsilon:
            break
        else:
            k_centres_old = k_centres
            
    print("Nombre de tours : ",nbTours)
    
    #On convertit maintenant les centres en RGB
    k_centres = [colorsys.hsv_to_rgb(k_centres[i][0]/360,k_centres[i][1]/100,k_centres[i][2]/100) for i in range(k_total)]
    #Passage en eniters
    k_centres = [tuple([int(k_centres[i][j]*255) for j in range(3)]) for i in range(k_total)]
    im.convert('RGB')
    return k_centres,assoc
```
